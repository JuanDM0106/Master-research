## **Deterministic parameters values aggregation**

```python
%%R
# Real expected irrigation data aggregation
values=read_excel("/content/Real expected irrigation.xlsx", sheet = "Real expected irrigation")
data<-data.frame(matrix(0, 26, 17))

for (i in 1:26){
  data[i,]<-c(colSums(values[1:14,3:16]), values[14,17:19])
  values<-values[-(1:14),]
  }
WriteXLS(data.frame(data), "R expected irrigation.xlsx")
```
